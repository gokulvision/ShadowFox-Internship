```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar", "Apr", "Apr"],
    "Product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "Sales": [50000, 30000, 45000, 35000, 60000, 40000, 55000, 42000],
    "Profit": [12000, 8000, 10000, 9000, 15000, 11000, 13000, 12000]
}

df = pd.DataFrame(data)
df
```

```python
print("Total Sales:", df["Sales"].sum())
print("Total Profit:", df["Profit"].sum())
```

```python
monthly = df.groupby("Month")["Sales"].sum()
monthly.plot(kind="bar")
plt.title("Monthly Sales")
plt.show()
```

```python
product = df.groupby("Product")["Sales"].sum()
product.plot(kind="pie", autopct="%1.1f%%")
plt.title("Product Sales Share")
plt.ylabel("")
plt.show()
```

```python
print("Insights:")
print("1. Laptop has higher sales than Phone.")
print("2. March recorded highest sales.")
print("3. Overall business is profitable.")
```
